```python
import pandas as pd
path="Resources/purchase_data.csv"
df= pd.read_csv(path)
#df.head()
```

```python
### Player Count
total_player_num=len(df["SN"].unique())
#print(total_player_num)
result=pd.DataFrame([total_player_num],columns=["Total Players"])
result.head()
```

```python
### Purchasing Analysis (Total)
unique_items=df["Item Name"].unique()
#unique_items
unique_items_number=len(unique_items)
#print(unique_items_number)
avg_purchase_price=df["Price"].mean()
#print(avg_purchase_price)
total_purchase_num=df["Item Name"].count()
#print(total_purchase_num)
total_avenue=df["Price"].sum()
#print(total_avenue)
result_array=[unique_items_number,avg_purchase_price,total_purchase_num,total_avenue]

# result=pd.DataFrame([{"Number of Unique Items":unique_items_number},
#                      {"Average Price":avg_purchase_price},
#                      {"Number of Purchases":total_purchase_num},
#                      {"Total Revenue":total_avenue}])
#This won't work because it's creating a 4x4 array with NaN values

result=pd.DataFrame(data=[result_array],columns=["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"])
#Need to do data=[result_array] other wise it will try to create a 4x1 data frame with 4 columns and returns an error
result["Average Price"]=result["Average Price"].map("${:.2f}".format)
result["Total Revenue"]=result["Total Revenue"].map("${:.2f}".format)
result.head()
```

```python
### Gender Demographics
male_purchase=df.loc[df["Gender"]=="Male"]
male_player_num=len(male_purchase["SN"].unique())
male_player_percent=male_player_num/total_player_num

female_purchase=df.loc[df["Gender"]=="Female"]
female_player_num=len(female_purchase["SN"].unique())
female_player_percent=female_player_num/total_player_num

other_purchase=df.loc[df["Gender"]=="Other / Non-Disclosed"]
other_player_num=len(other_purchase["SN"].unique())
other_player_percent=other_player_num/total_player_num

# print(male_player_num)
# print(male_player_percent)
# print(female_player_num)
# print(female_player_percent)
# print(other_player_num)
# print(other_player_percent)

result=pd.DataFrame({"Total Count":[male_player_num, female_player_num,other_player_num],
                     "Percentage of Players":[male_player_percent,female_player_percent,other_player_percent]
                    },
                    index=["Male","Female","Other / Non-Disclosed"]
                   )
result["Percentage of Players"]=result["Percentage of Players"].round(4)
result.head()
```

```python
### Purchasing Analysis (Gender)
male_purchase_count=male_purchase['SN'].count()
male_avg_price=male_purchase['Price'].mean()
male_total_value=male_purchase['Price'].sum()
male_avg_person=male_total_value/male_player_num
# print(male_purchase_count)
# print(male_avg_price)
# print(male_total_value)
# print(male_avg_person)

female_purchase_count=female_purchase['SN'].count()
female_avg_price=female_purchase['Price'].mean()
female_total_value=female_purchase['Price'].sum()
female_avg_person=female_total_value/female_player_num
# print(female_purchase_count)
# print(female_avg_price)
# print(female_total_value)
# print(female_avg_person)

other_purchase_count=other_purchase['SN'].count()
other_avg_price=other_purchase['Price'].mean()
other_total_value=other_purchase['Price'].sum()
other_avg_person=other_total_value/other_player_num
# print(other_purchase_count)
# print(other_avg_price)
# print(other_total_value)
# print(other_avg_person)

result=pd.DataFrame({"Gender":["Male", "Female","Other / Non-Disclosed"],
                     "Purchase Count":[male_purchase_count,female_purchase_count,other_purchase_count],
                     "Average Purchase Price":[male_avg_price, female_avg_price,other_avg_price],
                     "Total Purchase Value":[male_total_value, female_total_value,other_total_value],
                     "Avg Total Purchase per Person":[male_avg_person, female_avg_person,other_avg_person],
                    } 
                   )



result.set_index("Gender", inplace=True)
# Need inplace=True to get rid of the default 0123 index
result["Average Purchase Price"]=result["Average Purchase Price"].map("${:.2f}".format)
result["Total Purchase Value"]=result["Total Purchase Value"].map("${:.2f}".format)
result["Avg Total Purchase per Person"]=result["Avg Total Purchase per Person"].map("${:.2f}".format)
result.head()
```

```python
### Top Spenders
df_sn=df.groupby(["SN"])
df_sn_count=df_sn["SN"].count().sort_values(ascending=False)
#print(df_sn_count). This is the same as df["SN"].value_counts()
purchase_history={}
names=df_sn_count.index.tolist()
SN=[]
purchase_count=[]
avg_purchase_price=[]
total_purchase=[]

for i in range(5):
    df_names=df.loc[df["SN"]==names[i]]
    Name=names[i]
    purchase_history[Name]=0
    #print(df_names["Price"])
    purchase_history[Name]+=df_names["Price"]
    #print(purchase_history[Name])
    #In this case df_names["Price"] is a list and the whole list got stored in the corresponding key in purchase_history
    SN.append(Name)
    purchase_count.append(df_sn_count.iloc[i])
    total_purchase.append(purchase_history[Name].sum())
    avg_purchase_price.append(total_purchase[i]/purchase_count[i])
    
# print(SN)
# print(purchase_count)
# print(avg_purchase_price)
# print(total_purchase)

result=pd.DataFrame({"SN":SN,
                     "Purchase Count":purchase_count,
                     "Average Purchase Price":avg_purchase_price,
                     "Total Purchase Value":total_purchase
                    } 
                   )
result.set_index("SN", inplace=True)
result["Average Purchase Price"]=result["Average Purchase Price"].map("${:.2f}".format)
result["Total Purchase Value"]=result["Total Purchase Value"].map("${:.2f}".format)
result.head()
```

```python
### Most Popular Items
df_items=df.groupby(["Item Name"])
df_item_count=df_items["Item Name"].count().sort_values(ascending=False)
purchase_history={}
names=df_item_count.index.tolist()
item_name=[]
ID=[]
purchase_count=[]
avg_purchase_price=[]
total_purchase=[]
prices=[]
price=[]

for i in range(5):
    df_item=df.loc[df["Item Name"]==names[i]]
    df_item=df_item.reset_index()
    Name=names[i]
    purchase_history[Name]=0
    purchase_history[Name]+=df_item["Price"]
    prices.append(df_item["Price"])
    item_name.append(Name)
    ID.append(df_item.loc[0,"Item ID"])
    purchase_count.append(df_item_count.iloc[i])
    total_purchase.append(purchase_history[Name].sum())
    avg_purchase_price.append(total_purchase[i]/purchase_count[i])
    

for x in range(len(prices)):
    price.append(prices[x][0])
    # Need this to take out each price from the prices array which contains 
    # all entries for prices for the same item
    
    
# print(ID)
# print(item_name)
# print(purchase_count)
# print(avg_purchase_price)
# print(total_purchase)

result=pd.DataFrame({"Item\r\nID":ID,
                     "item Name":item_name,
                     "Purchase Count":purchase_count,
                     "Item Price":price,
                     "Total Purchase Value":total_purchase
                    } 
                   )

#Why can't I start a new line with Item\r\nID?
result.set_index("Item\r\nID", inplace=True)
result["Item Price"]=result["Item Price"].map("${:.2f}".format)
result["Total Purchase Value"]=result["Total Purchase Value"].map("${:.2f}".format)
result.head()
```
